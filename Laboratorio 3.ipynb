#### 1) ¿Cuántos Municipios tienen más de 100000 habitantes?

#### 2) Realice una gráfica de barras sobre la población de cada ciudad, ordenela de menor a mayo y responda:
a. ¿Cuál es la segunda ciudad más poblada?
b. ¿Qué posición ocupa Granada en el ranking de las ciudades más pobladas?

```python
poblacion_ciudad = df.groupby('Municipio')['Poblacion'].sum()
poblacion_ciudad = poblacion_ciudad.sort_values()
plt.barh(poblacion_ciudad.index, poblacion_ciudad.values)
plt.show()
segunda_ciudad_mas_poblada = poblacion_ciudad.index[-2]
print('La segunda ciudad más poblada es:', segunda_ciudad_mas_poblada)
ranking_de_granada = list(poblacion_ciudad.index).index('Granada') + 1
print('Granada ocupa la posición', ranking_de_granada, 'en el ranking de las ciudades más pobladas')
```

#### 3) ¿Cuántos municipios de Extremadura tienen más de 5000 habitantes?.

```python
municipios_extremadura = df[df['Provincia'] == 'Badajoz'] + df[df['Provincia'] == 'Cáceres']
num_municipios_mas_5000 = (municipios_extremadura['Poblacion'] > 5000).sum()

print('Hay', num_municipios_mas_5000, 'municipios de Extremadura con más de 5000 habitantes')
```

#### 4) ¿Cuál es el municipio situado más al Norte? (Usar el valor de la coordenada "Y" que representa la latitud en grados). Proporcione también la provincia a la que pertenece y su población.

```python
municipio_mas_norte = df.loc[df['Y'].idxmax()]
nombre_municipio = municipio_mas_norte['Municipio']
provincia = municipio_mas_norte['Provincia']
poblacion = municipio_mas_norte['Poblacion']

print('El municipio situado más al Norte es', nombre_municipio, 'en la provincia de', provincia, 'con una población de', poblacion, 'habitantes.')
```

#### 5) Obtenga la media, mediana, desviación estándar, valor máximo y valor mínimo de la población de los municipios de la provincia de Granada.

```python
df_granada = df[df['Provincia'] == 'Granada']
media = df_granada['Poblacion'].mean()
mediana = df_granada['Poblacion'].median()
desviacion_estandar = df_granada['Poblacion'].std()
valor_maximo = df_granada['Poblacion'].max()
valor_minimo = df_granada['Poblacion'].min()
print('Media:', media)
print('Mediana:', mediana)
print('Desviación estándar:', desviacion_estandar)
print('Valor máximo:', valor_maximo)
print('Valor mínimo:', valor_minimo)
```

#### 7) Seleccione al azar cincuenta municipios diferentes de  entre los diponibles en el archivo. Asegúrese de que no  se repitan. ¿Luego calcule el promedio de la población y la desviación estándar de esto 50 municipios?

```python
municipios_seleccionados = np.random.choice(df['Municipio'].unique(), size=50, replace=False)
df_seleccionados = df[df['Municipio'].isin(municipios_seleccionados)]
media = df_seleccionados['Poblacion'].mean()
desviacion_estandar = df_seleccionados['Poblacion'].std()
print('Media de población de los 50 municipios seleccionados:', media)
print('Desviación estándar de la población de los 50 municipios seleccionados:', desviacion_estandar)
```

#### 8) Dígame los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero

```python
import pandas as pd
from geopy.distance import geodesic

# Filtrar por municipios distintos a Madrid
df = df[df["Municipio"] != "Madrid"]

# Calcular la distancia a Madrid para cada municipio
df["Distancia a Madrid"] = df.apply(lambda row: geodesic((row["Y"], row["X"]), ("40.416775", "-3.703790")).km, axis=1)

# Municipio más cercano a Madrid
municipio_cercano = df.loc[df["Distancia a Madrid"].idxmin(), "Municipio"]
distancia_cercano = df.loc[df["Distancia a Madrid"].idxmin(), "Distancia a Madrid"]
print("Municipio más cercano a Madrid: {}, distancia: {} km".format(municipio_cercano, distancia_cercano))

# Municipio más lejano a Madrid
municipio_lejano = df.loc[df["Distancia a Madrid"].idxmax(), "Municipio"]
distancia_lejano = df.loc[df["Distancia a Madrid"].idxmax(), "Distancia a Madrid"]
print("Municipio más lejano a Madrid: {}, distancia: {} km".format(municipio_lejano, distancia_lejano))
```
